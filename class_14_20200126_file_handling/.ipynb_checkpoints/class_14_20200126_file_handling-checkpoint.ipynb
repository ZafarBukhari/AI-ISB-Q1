## Read Whole File:

```python
# with open (file_name) as <variable>
# this will read all lines from file <myFiles>

with open("abc.txt") as myFiles:
    content = myFiles.read() 
    print(type(content))
    print(content)
                            
```

## Read One Line from File:

```python
# with open (file_name) as <variable>
# this will read first line

with open("abc.txt") as myFiles:
    content = myFiles.readline() 
    print(type(content))
    print(content)
                            
```

## Read All Lines from File:

```python
# with open (file_name) as <variable>
# this will read all lines

with open("abc.txt") as myFiles:
    content = myFiles.readlines() 
    print(type(content))
    print(content)
                            
```

## Read only desired number of lines:

```python
# with open (file_name) as <variable>
# this will read first two lines

with open("abc.txt") as myFiles:
    content = myFiles.readlines()[:2]     # this will read two lines
    print(type(content))
    print(content)
                            
```

## How to write in a File:

```python
# with open (file_name) as <variable>
# this will read first two lines

with open("abc.txt") as myFiles:
    content = myFiles.readlines()[:2]     # this will read two lines
    myFiles.write("Pakistan Zindabad")
    print(type(content))
    print(content)
                            
```

```python
with open("abc2.txt", 'w') as f:
    content = f.readlines()
    print(content)
```

## Append Mode:

```python
with open("abc3.txt", 'a') as f:
    f.write("Hello World!")
```

```python
with open("abc3.txt", 'a') as f:
    f.write("\nHello Python!")   # this will write line at end of file.
```

```python
with open("abc4.txt", 'a+') as f:
    f.write("Hello PIAIC.")
    f.seek(0)
    c = f.readlines()
    print(c)
```

```python
with open("abc4.txt", 'a+') as f:
    f.write("\nHello Pakistan.")
    f.seek(0)
    c = f.readlines()
    print(c)
```

```python
with open("task1.txt", 'r') as f:
    contents = f.read()
    print(contents)
```

```python
with open("task1.txt", 'a') as f:
    f.write("\n\n[A,B,C]")
```

```python
with open("task1.txt", 'r+') as f:
    a = list(reversed(f.readlines()))
    b = list(reversed(['A','B','C']))
    c = a + b
    d = "".join(c)
    f.seek(0)
    f.write(d)
```

```python
with open("abc5.txt", "r+") as f:
    a = list(reversed(f.readlines()))
    b = list(reversed(["A","B","C"]))
    c = a + b
    d = "" .join(c)   
    f.seek(0)
#     print(d)
    f.write(d)
```
